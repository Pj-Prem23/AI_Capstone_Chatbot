```python
def importfile_datacleaning():

    #importing the file
    data = pd.read_csv(p1._Filepath,index_col=[0])
    
    #check for Null Values
    data.dropna(axis =0, inplace = True)
    
    #Drop Duplicates
    data.drop_duplicates(inplace=True)
    
    #Rename Columns
    data.rename(columns={'Data':'Date', 'Countries':'Country', 
                        'Industry Sector':'Industry', 
                        'Accident Level': 'Accident_Severity', 
                        'Potential Accident Level': 'Potential_Accident_Severity', 
                        'Genre':'Gender', 'Employee or Third Party':'Employee_or_Third_Party',
                        'Critical Risk': 'Critical_Risk'}, inplace=True)
    
    
    data["Description"] = data["Description"].apply(clean_description)
    data["Description"] = data["Description"].apply(remove_stopwords)
    return data
```

```python
def createStopwords():
    
        random.seed(6)
        nltk.download('stopwords')
        nltk.download('punkt')
        nltk.download('omw-1.4')
        nltk.download('wordnet')

        stop_words = stopwords.words('english')
        wordnet_lemmatizer = WordNetLemmatizer()     
               
        
        return stop_words
```

```python
#Identifying two letter words in the Description
def char_two(text):
    words = text.split(" ")
    words = [word for word in words if (len(word) <= 2 and word not in ' ')]
    return " ".join(words)

def clean_description(text):
    # lower case
    text= text.lower()
    # keep only alphabet
    text = re.sub('[^a-z ]','',text)
    # tokenize the text
    tokens = word_tokenize(text)
    # lemmatize 
    wordnet_lemmatizer = WordNetLemmatizer()
    tokens = [wordnet_lemmatizer.lemmatize(word) for word in tokens]
    return " ".join(tokens)

def remove_stopwords(text):
    stop_words = createStopwords()
    words = text.split(" ")
    charc = [word for word in words if (len(word) <= 2 and word not in ' ')]
    two_char = []
    for i in charc:
        split_i = i.split()
        for j in split_i:
            if j not in two_char:
                two_char.append(j)
    stop_words.extend(two_char)
   
    #remove stopwords 
    words = [word for word in words if word not in stop_words]
    return " ".join(words)
```

```python
def tfidf_vec():
    data = pd.read_pickle("./final_data.pkl")    
    
    tfidf_df = pd.DataFrame()
    for i in [1,2,3]:
        vec_tfidf = TfidfVectorizer(max_features=100, norm='l2', stop_words='english', lowercase=True, use_idf=True, ngram_range=(i,i))
        X = vec_tfidf.fit_transform(data['Description']).toarray()
        tfs = pd.DataFrame(X, columns=["TFIDF_" + n for n in vec_tfidf.get_feature_names()])
        tfidf_df = pd.concat([tfidf_df.reset_index(drop=True), tfs.reset_index(drop=True)], axis=1)
        
        encoded_df = pd.DataFrame()
        #Label Encoding
        encoded_df['Accident_Severity'] = LabelEncoder().fit_transform(data['Accident_Severity']).astype(np.int8)
        final_df =  encoded_df.join(tfidf_df.reset_index(drop=True))

    return final_df
```

```python
def buildNNModel():
    # define the model
    model = Sequential()
    model.add(Dense(64, input_dim=16, activation='relu', kernel_initializer='random_normal'))
    model.add(Dense(128, activation='relu', kernel_initializer='random_normal'))
    model.add(Dense(256, activation='relu', kernel_initializer='random_normal'))
    model.add(Dense(40, activation='relu', kernel_initializer='random_normal'))
    model.add(Dense(1, activation='linear'))

    # compile the nn model
    model.compile(loss='mse', optimizer=SGD(lr=0.001, momentum=0.9), metrics=['accuracy'])

    return model
```

```python
def buildNLPModel(max_tokens=2464,hidden_dim=16):
    model = Sequential()
    model.add(Dense(32,input_shape=(max_tokens,)))
    model.add(Dense(hidden_dim))
    model.add(Dropout(0.5))
    model.add(Dense(5, activation='softmax'))
    model.compile(optimizer='rmsprop',loss='categorical_crossentropy',metrics=['accuracy'])
    
    return model
```
